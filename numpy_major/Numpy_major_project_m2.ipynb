```python
import numpy as np
```

```python
def get_input(name):
  while True:
    try:
      arr=input(f"Enter elements of array {name} separated by spaces: ")
      return np.array(list(map(float,arr.split())))
    except ValueError:
      print("Invalid input.Please enter numeric values")

def arthimetic(a,b):
  if a.size!=b.size:
    print("Arrays must be of same length")
    return
  print("\n Arthimetic Operations")
  print("Addition:",a+b)
  print("Subtraction: ",a-b)
  print("Multiplication: ",a*b)
  print("Division: ",a/b)
  print("Power: ",a**b))

def statistical_operations(a):
  print("\n statistical Operations")
  print("Mean:",np.mean(a))
  print("Median:", np.median(a))
  print("Standard Deviation: ",np.std(a))
  print("Variance:",np.var(a))
  print("Minimun: ",np.min(a))
  print("Maximun: ",np.max(a))
  print("Sum: ",np.sum(a))
  print("Product: ",np.prod(a))
  print("Shape: ",a.shape)
  print("Size: ",a.size)
  print("Data type: ",a.dtype)

def array_manipulation(a):
  print("\n Array Manipulation")
  try:
    rows=int(input("Enter no.of rows for reshape: "))
    if rows>0:
      if a.size%rows==0:
        reshaped=a.reshape(rows,-1)
        print("Reshaped array:\n",reshaped)
      else:
        print("Cannot reshape array into that number of rows")
  except ValueError:
    print("Invalid reshape input")

  print("sorted array: ",np.sort(a))
  print("Reverse Sorted array: ",np.sort(a)[::-1])
  print("unique values: ",np.unique(a))
  print("First element: ",a[0])
  print("Last Element: ",a[-1])
  print("Values greater than mean:",a[a>np.mean(a)])

print("\nNumpy Mathematical Calculator:")

while True:
    print("\nChoose an operation")
    print("1.Basic Arthimetaic Operations")
    print("2.statistical Operations")
    print("3. Array Manipulation")
    print("4.Exit")

    choice=input("Enter your choice(1-4): ")

    if choice=="1":
      print("you chose Basic arthimetic operations")
      a=get_input("A")
      b=get_input("B")
      arthimetic(a,b)

    elif choice=="2":
      print("you chose statistical operations")
      a=get_input("A")
      statistical_operations(a)

    elif choice=="3":
      print("You chose array manipulation")
      a=get_input("A")
      array_manipulation(a)

    elif choice=="4":
      print("Exit")
      break

    else:
      print("Invalid choice.")



```
